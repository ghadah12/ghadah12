# Classification

Classification is a type of supervised learning which is why labeled data is required. "Genetics" has been used as the class label. Additionally, encoding has been done in order to convert categorial attributes into numeric ones.

```python
label_encoder = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = label_encoder.fit_transform(df[col])
```

The code initializes a LabelEncoder object to encode categorical variables into numerical values. It then iterates through each column of the DataFrame, checks if the datatype is 'object', and if so, applies label encoding to convert categorical values into numerical ones.

```python
X = df.drop(columns=["Genetics", "Id"])
y = df["Genetics"]
```

The code prepares the feature matrix (X) by dropping the columns "Genetics" and "Id" from the DataFrame then assigns the column "Genetics" as the target variable (y) for the model.

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print(cm)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No', 'Yes'])
disp.plot(cmap=plt.cm.Reds)
plt.show()
```

```python
partition_sizes = [0.1, 0.2, 0.3]

criterions = ["entropy", "gini"]

results = {}

for size in partition_sizes:
    for criterion in criterions:
        
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size, random_state=42)

        clf = DecisionTreeClassifier(criterion=criterion, random_state=42)
        
        clf.fit(X_train, y_train)
        
        y_pred = clf.predict(X_test)
        
        accuracy = accuracy_score(y_test, y_pred)
        
        results[(size, criterion)] = accuracy

for (size, criterion), accuracy in results.items():
    print(f"Partition Size: {size}, Criterion: {criterion}, Accuracy: {accuracy}")

sizes_entropy = [size for size, criterion in results if criterion == "entropy"]
accuracies_entropy = [results[(size, "entropy")] for size in sizes_entropy]
sizes_gini = [size for size, criterion in results if criterion == "gini"]
accuracies_gini = [results[(size, "gini")] for size in sizes_gini]

plt.figure(figsize=(10, 6))
plt.plot(sizes_entropy, accuracies_entropy, marker='o', label='Entropy')
plt.plot(sizes_gini, accuracies_gini, marker='o', label='Gini')
plt.title('Accuracy vs Partition Size')
plt.xlabel('Partition Size')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

Using Entropy the accuracy increases sharply when the partition size lies between 0.100 to 0.200 and decreases sharply from partition size 0.200 and above.
Using Gini the accuracy increases when the partition size lies between 0.100 to 0.200 and decreases gradually from partition size 0.200 and above.

# Clustering

In unsupervised learning, like clustering, there's no need for labeled data. Thus, we'll omit the "Selector" class label. Since our dataset contains only numeric attributes, we won't require any additional encoding. Next, we'll standardize the attributes to ensure their scales are comparable, preventing any single attribute from unduly influencing the clustering process due to varying magnitudes.

## the choice for the three different sizes of K-means clustering

For selecting the three different sizes for K-means clustering, we'll rely on validation methods to guide our decision. Once determined, we'll use these sizes to conduct the clustering. Our aim is to ensure diversity in size selection: one will be relatively large, another moderate, and the last one small. This approach aims to cover a wide range of potential cluster configurations.

# silhouette method:

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Assuming df_scaled is defined earlier in the code
strokes = df_scaled

# Function to calculate silhouette score for a given number of clusters
def get_silhouette_score(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(data)
    silhouette_avg = silhouette_score(data, labels)
    return silhouette_avg

# Finding the optimal number of clusters using the silhouette method
range_n_clusters = range(2, 11)  # Change the range as needed
silhouette_scores = [get_silhouette_score(strokes, n) for n in range_n_clusters]

# Plotting the results
plt.figure(figsize=(4, 3)) 
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Avg Silhouette Score')
plt.title('Silhouette Method for k-Mean Clusters')
plt.grid(True)
plt.show()
```

Silhouette analysis suggests that 2 clusters are optimal for the dataset as they yield the highest silhouette score among the tested range (2 to 10 clusters), indicating the best structure for segmentation.
