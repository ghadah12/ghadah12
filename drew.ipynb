```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.feature_selection import VarianceThreshold 
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from scipy import stats
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

data= pd.read_csv('Predict Hair Fall.csv')

df=pd.DataFrame(data)
```

```python
df.replace("No Data", np.nan, inplace=True)

missing = df.isna().sum()
print(missing.sum())

Data_cleand = df.dropna()

print(Data_cleand)
```

```python
le = LabelEncoder()
df['Genetics'] = le.fit_transform(df['Genetics'])

df['Hormonal Changes'] = le.fit_transform(df['Hormonal Changes'])

#df['Poor Hair Care Habits'] = le.fit_transform(df['Poor Hair Care Habits'])
#df['Weight Loss'] = le.fit_transform(df['Weight Loss'])

df['Environmental Factors'] = le.fit_transform(df['Environmental Factors'])

df['Smoking'] = le.fit_transform(df['Smoking'])

df['Medical Conditions'] = le.fit_transform(df['Medical Conditions'])

df['Medications & Treatments'] = le.fit_transform(df['Medications & Treatments'])

#df['Nutritional Deficiencies'] = le.fit_transform(df['Nutritional Deficiencies'])

df['Stress'] = le.fit_transform(df['Stress'])

print(df)
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
X = df.select_dtypes(include=['number']).drop(columns=['Poor Hair Care Habits'])  # Select numerical features
y = df['Poor Hair Care Habits']  # Target variable

# Assuming df is your DataFrame and 'target_column' is the target variable
X = df.drop(columns=['Genetics'])  # Features
y = df['Genetics']  # Target variable

# SelectKBest with chi-squared test
best_features = SelectKBest(score_func=chi2, k=2)  # Select top 2 features
fit = best_features.fit(X, y)

# Get scores and corresponding feature names
feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': fit.scores_})
feature_scores = feature_scores.sort_values(by='Score', ascending=False)

# Print top 2 features
print("Top 2 features:")
print(feature_scores.head(2))
```
